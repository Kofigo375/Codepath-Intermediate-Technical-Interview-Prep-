## Advanced concepts 

```python
## iterating over the indeces and characters in a string
my_string = 'code'
for ind, cha in enumerate(my_string):
    print(ind, cha)
```

```python
##  using enumerate and putting in a start value
cereals = ['cheerios', 'fruity peebles', 'cocoa puffs']
for count, cereal in enumerate(cereals, start=1):
    print(count, cereal)
```

```python
# zipping 2 lists
names = [ 'Alice', 'Bob', 'Charlie']
ages = [25, 30, 35]

zipped = zip(names, ages)
print(zipped)
```

```python
# list comprehension 
nums = [1,2,3,4,5]
doubled = [ num ** 2 for num in nums]
print(doubled)
```

```python
## adding conditions to list comprehensions 
words = ["I", "Love", "Codepath"]
result = []

for word in words:
    if len(word) > 5:
         result.append(word)
    
print(result)
```

```python
words = ["I", "Love", "Codepath"]

result = [ word for word in words if len(word) > 5]
print(result)
```
