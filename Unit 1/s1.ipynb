```python
# Problem 12 : Shuffle 

""" 
I"m given a list that  some even number of elements 
I should split the and piar them 
"""


def shuffle(cards):
    n = len(cards) // 2
    shuffled = []
    for i in range(n):
        shuffled.append(cards[i])
        shuffled.append(cards[i+n])
    return shuffled
```

```python
cards = ['Joker', 'Queen', 2, 3, 'Ace', 7]
shuffle(cards)

```

```python
# Problem 2 Bouncy, Flouncy, Trouncy, Pouncy 

# understanding 
""" 
we have some strings that corresponds to some operations , incre or decre
we have a variable which is initially 1 
write a func so that when i'm given a list of strings which corresponds to some operations 
I will return the final value of the of the variable tigger after performing the correpond operations on it 
"""


# plain 
""" 
initialize the tigger varaible 
iterate through the list 
if the string is bouncy or flouncy , I will increament the value of tiger by 1
if the string is trouncy or i will decrease the tigger by 1 

"""

# implement 

def final_value_after_operations(operations):
    tigger = 1
    for operation in operations:
        if operation == "bouncy" or operation ==  "flouncy":
            tigger +=1 
        elif operation == "trouncy" or operation == "pouncy":
            tigger -= 1 
    return tigger
```
