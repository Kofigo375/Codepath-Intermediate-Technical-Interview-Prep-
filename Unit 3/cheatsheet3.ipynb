```python
## implementing a stack 

stack = [] 

#push items onto the stack
stack.append(2)
stack.append(1)


# pop items onto the stack 
popped_items = stack.pop()

print(popped_items)
print(stack)


```

```python
## queue implementation 
from collections import  deque

queue = deque()

queue.append(1)
queue.append(3)
queue.append(8)

print(queue)
```

```python
removed_element = queue.popleft()
print(removed_element)
```
