```python
## Problem 1 : Post Format Validator 
## given a string , check if the tags in it are correctly formated

def is_valid_post_format(posts):
    # initialize  empty stack to keep track of the opening tags 
    stack = []
    
    # initialize the matcing pairs in a dictionary
    # for this closing tag this is the corresponding opening tag 
    pairs = {
        ')': '(',
        ']': '[',
        '}': '{'
    }
    
    # iterate through the posts 
    for char in posts: 
    ## you see an opening tag push it onto the stack to later check its match 
        if char in '([{':
            stack.append(char)
    #  if it's a closing tag 
        elif char in ')]}':
        # check if the stack is not empty , if empty it means a closing tag is going in without a pair 
            if not stack:
                return False
        # if it is not empty check the opening tag on top if it is a pair 
            top = stack.pop()
            if top != pairs[char]:
                return False 
        
        # if it does not match return false 
        
    # after going through the whole posts characters 
    # return True if the stack is empty : means all chars were matched and poped 
    # return false if stack is not empty : means some chars did not have a match so they were not poped  
    return len(stack)==0    
    
    
```

```python
print(is_valid_post_format("()"))
print(is_valid_post_format("()[]{}")) 
print(is_valid_post_format("(]"))
```

```python
## Problem 2: Reverse User Comments Queue

## Understand:  when I'm given a queue in the form of a list of strings as comments , I should reverse the order using a stack ?

##  Plan :
    # use 2 opposite pointers 
    # iterate through the queue 
    # check if the left pointer is less than right pointer meaning ( the left pointer has not moved past the right ponter )
    # at every iteration I swap the values of the pointers and 
    # increase the lower pointer and decrease the higher pointer 
    
    
def reverse_comments_queue(comments):
    left_pointer = 0
    right_pointer = len(comments) - 1
    
    for char in comments:
        if left_pointer < right_pointer:
            comments[left_pointer], comments[right_pointer] = comments[right_pointer], comments[left_pointer] 
            left_pointer += 1
            right_pointer -= 1
            
    return comments
            
    
## corection : use a stack and push pop approach 
# Plan :
    ## initialize an empty stack 
    ## iterate through comments in input list 
        #   push each  comment onto the stack
    ## intialize an empty list to hold reversed comment
    ## pop element from the stack to the reversed list while the stack is not empty 
    # once the stack is empty return the reversed list 
    
## the push and pop ( Last in first out (LIFO)) guarantees the reversal 

def reverse_comments_queue(comments):
    stack = []
    for comment in comments:
        stack.append(comment)
    reversed_comments = []
    
    while stack:
        reversed_comments.append(stack.pop())
    
    return reversed_comments

    
```

```python
print(reverse_comments_queue(["Great post!", "Love it!", "Thanks for sharing."]))
```

```python
print(reverse_comments_queue(["First!", "Interesting read.", "Well written."]))
```
